# **LIVERPOOLS ANALYTICS CHALLENGE**

### **Key Factors in Liverpools Attack:**
1) They keep creating chances, they dont let the defense settle at all. 
<br>2)They love penetrating the defense from the flanks rather than the center.
<br>3) They play high-line defense which helps the fullbacks combine with attack so easily.
<br>4) Their organisation is so good that the 2nd ball is always won by them.
<br>5) At a any given time of cross there are always more than 5 players in the Penalty area waiting to cause danger.
<br>6) They are very good in handling Aerial Balls.

### **Key Players in Liverpools Attack**:
1) **Firmino** - Being Centre Forward of the team, very often firmino is found creating chances rather than scoring goals. This is because Firmino plays deep attracting either of the centre back towards him which creates space for Mane or Salah to make surprise runs. Firmino also makes intelligent and timing runs which helps him to create chances.

2) **The fullbacks**- The fullbacks are the heart of Liverpools Attack. As they play a very high-line defense, the fullbacks play like the wingers which helps the 3 forwards to stay infront of the post increasing the chance of scoring.

```
from matplotlib import pyplot as plt
import pandas as pd

import sys, os
sys.path.insert(0, os.path.abspath('../scripts/'))
import footyviz

#Code from Lastrowmasters 

data = pd.read_csv('/content/drive/My Drive/Last-Row-master/datasets/positional_data/liverpool_2019.csv', index_col=('play', 'frame'))
play='Liverpool [2] - 1 Chelsea'
df=data.loc[play] 
from moviepy import editor as mpy
from moviepy.video.io.bindings import mplfig_to_npimage

def draw_frame_x(df, t, fps, voronoi=False):
  fig,ax,dfFrame = footyviz.draw_frame(df, t=t, fps=fps)
  if voronoi:
      fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
  image = mplfig_to_npimage(fig)
  plt.close()
  return image    

def make_animation(df, fps=20, voronoi=False):
    #calculated variables
  length=(df.index.max()+20)/fps
  clip = mpy.VideoClip(lambda x: draw_frame_x(df, t=x, fps=fps, voronoi=voronoi), duration=length-1).set_fps(fps)
  return clip

clip=make_animation(df) 

clip_voronoi = make_animation(df, voronoi=True)
composite_clip = mpy.CompositeVideoClip([clip, clip_voronoi.resize(0.2).set_position((10,-1))])
clip.rotate(90).crop(y1=100, y2=500).ipython_display()



```
